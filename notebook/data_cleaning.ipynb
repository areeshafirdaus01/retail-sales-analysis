```python
import pandas as pd
df= pd.read_excel('Clothing Sales Messy Data.xlsx')
df= df.drop_duplicates()
df['unitPrice']=df['unitPrice'].fillna(df['totalAmount']/df['quantity'])
df['costPrice']=df['costPrice'].fillna(df['totalCost']/df['quantity'])
df['totalCost']=df['totalCost'].fillna(df['unitPrice']*df['quantity'])
df['totalAmount']=df['totalAmount'].fillna(df['unitPrice']*df['quantity'])
df = df.dropna(subset=['unitPrice', 'costPrice','totalAmount','totalCost','quantity'])
df["saleDate"] = pd.to_datetime(df["saleDate"], format='mixed', dayfirst=True, errors='coerce')
df["profit"]= df['totalAmount']- df['totalCost']
df['profit'].sum()
df.head()
df.to_csv('Clothing Sales Cleaned Data.csv', index=False)
```
